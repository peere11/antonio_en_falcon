```python
import numpy as np

#Coordenadas (Longitud, Latitud) de ciudades visitadas para calcular distancias

#Cuatro vientos
CV = (-3.794700, 40.376580)
#Moncloa
MO = (-3.719410, 40.434990)
#Extremadura (Montfragüe)
EX = (-5.943268, 39.804100)
#Torrejón de Ardoz
TO = (-3.475337, 40.453298)
#León
LE = (-5.572589, 42.596897)
#Ourense
OU = (-7.841412, 42.338646)
#Castellón
CA = (-0.096917, 39.974161)

#Velocidad de crucero Falcon i Superpuma (km/h)
v_falcon = 950
v_puma = 278

v_coche = 90

#Consumo del Falcon i el SúperPuma (en l/h) 
consumo_falcon = 1313
consumo_puma = 600

#Emisiones CO2 (kg CO2 / litro combustible)
CO2_em = 2.527

def co2(coord1, coord2, transporte):
    
    #Función que calcula la emisión de CO2 entre puntos (coord1,coord2) para Falcón o Helicóptero
    
    #Conversion de latitud i longitud a kilometro (km/º)
    
    lat_km = 110.574
    
    #Latitud media
    lat_med = (coord1[1] + coord2[1])/2
    lon_km = 111.320*np.cos(lat_med*2*np.pi/360)
    
    #Distancia recorrida
    dist = np.sqrt(((coord1[0] - coord2[0])*lon_km)**2 + ((coord1[1] - coord2[1])*lat_km)**2)
    
    if transporte=='Falcon':
        #Horas viage
        horas = dist/v_falcon
        #Consumo total combustible
        cons = horas*consumo_falcon
        
    elif transporte=='Puma':
        #Horas viage
        horas = dist/v_puma
        #Consumo total combustible
        cons = horas*consumo_puma
        
    elif transporte=='Gasolina':
        cons = 2.35*dist/12
    
    #Emisiones CO2 en el trayecto
    
    return dist,cons*CO2_em

#Listado de trayectos entre 18 i 20 de julio

trayectos = [[CV,MO, 'Puma'], [MO,EX, 'Puma'], [EX, CV, 'Puma'],
             [TO,EX,'Falcon'], [EX, TO, 'Falcon'],
            [CV,TO,'Puma'], [TO,MO,'Puma'],[MO,CV,'Puma'],
            [CV,LE,'Puma'],
            [CV,MO,'Puma'], [MO,TO,'Puma'], [TO,CV,'Puma'],
            [TO,LE,'Falcon'],
            [LE,OU,'Puma'],[OU,LE,'Puma'],
            [LE,TO,'Falcon'],
            [CV,TO,'Puma'], [TO,MO,'Puma'], [MO,CV,'Puma'],
            [CV,MO,'Puma'], [MO,TO,'Puma'], [TO,CV,'Puma'],
            [TO,CA,'Falcon'],[CA,TO,'Falcon']]

#Calculando Emisiones de CO2 para cada trayecto

emisiones_trayecto = []

for viaje in trayectos:
    
    kilosCO2 = co2(viaje[0],viaje[1],viaje[2])
    
    emisiones_trayecto.append(kilosCO2[1])
    
total_emisiones = np.sum(emisiones_trayecto)
    
print('Antonio ha consumido {:0.02f} kg de CO2 para hacerse una foto i culpar al cambio climático de la existencia de incendios en España.\n'.format(total_emisiones))

#Comparando con el ciudadano medio

#Emision co2 por litro de gasolina (kg CO2 por litro)
co2_gasolina = 2.30
#Consumo medio de coche de gasolina (km/l)
km_por_litro = 12
#Distancia anual media recorrida por un español
dist_anual = 12000

#Emision de CO2 anual media del español

co2_anual_esp = co2_gasolina*dist_anual/km_por_litro
print('Teniendo en cuenta que un español conduciendo un coche de gasolina emite de media unos {:0.02f} kg de CO2 anuales.\n'.format(co2_anual_esp))
print('Antonio ha consumido en tres dias lo que consumen unos {:0.01f} conductores en un año.\n'.format(total_emisiones/co2_anual_esp))
```
